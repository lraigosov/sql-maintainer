# Glosario SQL

**Objetivo**: referencia rápida de conceptos clave del curso.

---

## A
- **Agregación**: operación que combina múltiples filas en un resultado único (SUM, COUNT, AVG, etc.).
- **Alucinación (IA)**: generación de contenido SQL incorrecto o inventado por IA sin validación.

## C
- **CTE (Common Table Expression)**: subconsulta nombrada que mejora legibilidad; se define con WITH.
- **Cardinalidad**: número de filas únicas en una columna o resultado.
- **Clave sustituta**: identificador artificial (surrogate key) sin significado de negocio.
- **Correlacionada (subconsulta)**: subconsulta que referencia columnas de la consulta externa.

## D
- **Dimensión**: tabla descriptiva (clientes, productos, fechas) en modelo estrella.
- **DENSE_RANK()**: función de ventana que asigna ranking sin saltos en caso de empates.

## E
- **EXISTS**: operador para verificar existencia de filas en subconsulta.
- **ETL**: Extract, Transform, Load; proceso de integración de datos.

## F
- **Función de ventana (window function)**: cálculo sobre conjunto ordenado sin colapsar filas (RANK, LAG, LEAD, SUM OVER).
- **Fact table (tabla de hechos)**: almacena transacciones/eventos con métricas cuantificables.

## G
- **Granularidad**: nivel de detalle de datos (ej: por día, por hora, por transacción).
- **Gobierno de datos**: políticas y controles para calidad, seguridad y uso de datos.

## I
- **Índice compuesto**: índice sobre múltiples columnas para optimizar filtros combinados.
- **Índice cubridor (covering index)**: incluye todas las columnas requeridas en una consulta, evitando búsquedas adicionales.

## J
- **JOIN**: operación para combinar filas de dos tablas según condición (INNER, LEFT, RIGHT, FULL).

## K
- **KPI (Key Performance Indicator)**: métrica crítica de negocio (ej: margen, ticket promedio).

## L
- **LAG/LEAD**: funciones de ventana que acceden a fila previa/siguiente en conjunto ordenado.

## M
- **Margen**: diferencia entre precio de venta y costo; métrica de rentabilidad.
- **Modelo estrella (star schema)**: arquitectura con tabla de hechos central y dimensiones satelitales.

## N
- **NULL**: ausencia de valor; requiere manejo especial (IS NULL, ISNULL, COALESCE).
- **NULLIF**: devuelve NULL si dos expresiones son iguales; útil para evitar división por cero.

## O
- **OVER()**: cláusula que define ventana para funciones analíticas (PARTITION BY, ORDER BY).
- **Outlier**: valor atípico o fuera de rango esperado.

## P
- **Partición (partition)**: división lógica o física de tabla grande por criterio (fecha, región).
- **Plan de ejecución**: representación del motor SQL de cómo procesará una consulta.
- **Prompting**: arte de formular instrucciones claras para IA generativa.

## R
- **RANK()**: función de ventana que asigna posición ordinal (con saltos en empates).
- **RFM**: Recency, Frequency, Monetary; técnica de segmentación de clientes.
- **ROW_NUMBER()**: asigna número secuencial único a cada fila en ventana.

## S
- **Sargable**: condición de filtro que permite uso eficiente de índices (sin funciones sobre columna indexada).
- **Segmentación**: clasificación de entidades (clientes, productos) en grupos homogéneos.
- **Subconsulta**: SELECT dentro de otro SELECT; puede ser escalar, de columna o de tabla.

## T
- **Ticket promedio**: ingresos totales divididos por número de transacciones.
- **Transacción**: unidad lógica de trabajo (ACID: Atomicity, Consistency, Isolation, Durability).

## U
- **UNION**: combina resultados de dos SELECT eliminando duplicados; UNION ALL conserva todos.

## V
- **Vista (view)**: consulta guardada como objeto; puede simplificar acceso o seguridad.

## W
- **Window function**: ver *Función de ventana*.
- **WITH**: cláusula para definir CTE.
