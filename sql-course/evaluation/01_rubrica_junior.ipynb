# 6.1 Rúbrica Nivel Junior

- Objetivo: evaluar fundamentos y proyecto 1.9.
- Criterios: sintaxis, joins, agregaciones, calidad de datos básica.
- Entregable: checklist con evidencias (queries).

---

## Criterios de Evaluación

### 1. Sintaxis y SELECT básico (20 puntos)
- [ ] Selecciona columnas específicas (no SELECT *).
- [ ] Usa alias descriptivos.
- [ ] Aplica ORDER BY correctamente.
- [ ] Filtra con WHERE (operadores básicos).

### 2. JOIN y relaciones (25 puntos)
- [ ] Identifica claves primarias y foráneas.
- [ ] Usa INNER JOIN correctamente.
- [ ] Usa LEFT JOIN para detectar faltantes.
- [ ] Evita productos cartesianos.

### 3. Agregaciones (20 puntos)
- [ ] SUM, COUNT, AVG correctos.
- [ ] GROUP BY con columnas adecuadas.
- [ ] HAVING para filtrar grupos.
- [ ] Métricas calculadas (margen, ticket promedio).

### 4. Subconsultas (15 puntos)
- [ ] Subconsulta en SELECT o WHERE.
- [ ] EXISTS para validación de existencia.
- [ ] Evita subconsultas innecesarias.

### 5. Calidad de datos (10 puntos)
- [ ] Detecta nulos.
- [ ] Valida rangos (ej: descuento > 50%).
- [ ] Identifica productos sin ventas.

### 6. Proyecto 1.9 (10 puntos)
- [ ] Entrega al menos 5 consultas funcionales.
- [ ] Queries responden preguntas de negocio.
- [ ] Código limpio y comentado.

---

## Escala

- **90-100**: Excelente. Listo para Level02.
- **75-89**: Bueno. Revisar joins y agregaciones.
- **60-74**: Aceptable. Reforzar fundamentos antes de avanzar.
- **<60**: Requiere reestudio de Level01.
