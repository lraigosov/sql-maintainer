# 6.3 Rúbrica de Evaluación Nivel Senior

Esta rúbrica evalúa competencias de ingeniero de datos senior (Level03) con foco en diseño de sistemas, performance, seguridad y gobernanza.

---

## Criterios de Evaluación (100 puntos)

### 1. Diseño de Arquitectura de Datos (20 puntos)
- **16-20 pts**: diseña modelos dimensionales complejos (snowflake, galaxy), SCD Tipo 2, particiona tablas estratégicamente
- **12-15 pts**: diseña star schema funcional pero sin considerar evolución temporal (SCD), particionado básico
- **8-11 pts**: modelo relacional sin optimización para analítica, no aplica SCD ni particiones
- **0-7 pts**: diseño incoherente, no escala, problemas de integridad referencial

### 2. Optimización y Performance (25 puntos)
- **20-25 pts**: usa índices compuestos, columnstore, particiones, interpreta planes de ejecución, aplica hints estratégicamente
- **15-19 pts**: crea índices básicos, revisa planes pero no optimiza, desconoce columnstore
- **10-14 pts**: no usa índices, consultas lentas sin análisis de bottlenecks
- **0-9 pts**: anti-patrones de performance (scan de tablas masivas, cursores innecesarios)

### 3. Procedimientos Almacenados y Automatización (20 puntos)
- **16-20 pts**: SPs modulares con manejo robusto de errores (TRY/CATCH), parámetros validados, logging de auditoría, transacciones ACID
- **12-15 pts**: SPs funcionales pero sin manejo de errores, validaciones mínimas, transacciones parciales
- **8-11 pts**: SPs monolíticos, sin transacciones, no maneja rollback
- **0-7 pts**: SPs con SQL injection vulnerabilities, sin manejo de errores

### 4. Seguridad y Compliance (15 puntos)
- **12-15 pts**: implementa roles granulares, RLS (Row-Level Security), auditoría de accesos, encripta datos sensibles
- **9-11 pts**: roles básicos pero sin RLS, auditoría manual, no encripta PII
- **6-8 pts**: permisos monolíticos (db_owner a todos), sin auditoría
- **0-5 pts**: vulnerabilidades críticas de seguridad, acceso sin control

### 5. Observabilidad y Mantenimiento (10 puntos)
- **8-10 pts**: monitorea fragmentación, automatiza rebuild/reorganize, mantiene estadísticas actualizadas, configura alertas de performance
- **6-7 pts**: mantenimiento manual reactivo, no monitorea proactivamente
- **3-5 pts**: índices nunca reconstruidos, estadísticas desactualizadas
- **0-2 pts**: no realiza mantenimiento, BD degradada

### 6. Gobernanza y Documentación (10 puntos)
- **8-10 pts**: documenta DDL, DML crítico, genera diccionario de datos, versionamiento de esquemas, políticas de retención
- **6-7 pts**: documentación parcial, sin diccionario de datos, versionamiento informal
- **3-5 pts**: sin documentación, cambios de esquema sin registro
- **0-2 pts**: código sin comentarios, imposible mantener

---

## Escala de Calificación
- **90-100**: Excelente - Arquitecto de datos senior, listo para liderar proyectos críticos
- **75-89**: Bueno - Sólido en diseño e implementación, necesita profundizar en gobernanza
- **60-74**: Aceptable - Conoce conceptos avanzados pero falta experiencia práctica en sistemas complejos
- **<60**: Necesita refuerzo - Revisar Level03 y ganar experiencia en proyectos reales

---

## Proyecto de Evaluación
**Caso práctico**: Diseñar e implementar sistema de data warehouse para e-commerce con:
1. Modelo dimensional con 5 dimensiones (clientes, productos, tiempo, geografía, promociones) y 3 hechos (ventas, devoluciones, inventario)
2. SCD Tipo 2 en dim_productos (historial de precios)
3. Particionado por mes en fact_ventas (últimos 36 meses)
4. Índices columnstore en fact tables
5. SP de carga incremental con manejo de errores y auditoría
6. Roles de seguridad (analista_lectura, ingeniero_escritura, admin_full)
7. Job automatizado de mantenimiento (rebuild índices si frag >30%)
8. Dashboard de observabilidad (top queries lentas, fragmentación, espacio en disco)
9. Documentación técnica completa (DER, diccionario de datos, runbook de mantenimiento)

**Entregables**:
- Scripts DDL completos
- SPs de ETL y mantenimiento
- Consultas analíticas optimizadas (<3s en 100M registros)
- Documentación en Markdown
