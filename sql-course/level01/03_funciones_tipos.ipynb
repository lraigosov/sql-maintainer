- Objetivo: usar funciones escalares (texto, fecha, num√©ricas) y conversiones.
- Prerrequisitos: conocer columnas de dim_productos y dim_clientes.
- Ejercicios: UPPER, YEAR, ROUND, DATEDIFF.
- Reto: formateo de cadenas y c√°lculo de antig√ºedad.
- Errores comunes: conversiones impl√≠citas, funciones en WHERE indexado.

```python
-- May√∫sculas en nombres de productos
SELECT producto_id, UPPER(nombre) AS nombre_mayus, categoria
FROM dbo.dim_productos;
```

```python
-- A√±o de alta de clientes y conteo
SELECT YEAR(fecha_alta) AS anio_alta, COUNT(*) AS clientes_registrados
FROM dbo.dim_clientes
GROUP BY YEAR(fecha_alta)
ORDER BY anio_alta;
```

üü¢ Ejercicio: Extraer mes de alta y agrupar clientes por segmento y mes.

```python
SELECT MONTH(fecha_alta) AS mes_alta, segmento, COUNT(*) AS clientes
FROM dbo.dim_clientes
GROUP BY MONTH(fecha_alta), segmento
ORDER BY mes_alta, segmento;
```

üü† Ejercicio: Redondear margen promedio de productos a 2 decimales.

```python
SELECT ROUND(AVG(precio_unitario - costo_unitario),2) AS margen_promedio
FROM dbo.dim_productos;
```

üî¥ Reto: Diferencia en d√≠as entre primera y √∫ltima fecha de venta registrada.

```python
SELECT DATEDIFF(day, MIN(fecha), MAX(fecha)) AS dias_span
FROM dbo.fact_ventas;
```
