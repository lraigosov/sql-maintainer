üü† Ejercicio: Productos activos ordenados por margen (precio - costo).

```python
SELECT producto_id, nombre, (precio_unitario - costo_unitario) AS margen
FROM dbo.dim_productos
WHERE activo = 1
ORDER BY margen DESC;
```
