## Paso 2: Configurar la Conexi√≥n

Vamos a usar `sqlalchemy` para crear una cadena de conexi√≥n. Modifica las variables a continuaci√≥n seg√∫n tu configuraci√≥n.

```python
import sqlalchemy
import pandas as pd

# CONFIGURACI√ìN - EDITA ESTO
SERVER = 'localhost'  # o tu servidor, ej: 'DESKTOP-XYZ\SQLEXPRESS'
DATABASE = 'SQLCourseDB' # La BD donde corriste dataset_setup.sql
DRIVER = 'ODBC Driver 17 for SQL Server' # Verifica tus drivers ODBC instalados
TRUSTED_CONNECTION = 'yes' # 'yes' para autenticaci√≥n de Windows, 'no' para usuario/pass

# Si usas autenticaci√≥n SQL (usuario/pass), llena esto y pon TRUSTED_CONNECTION = 'no'
USERNAME = 'sa'
PASSWORD = 'TuPassword'

# Construcci√≥n del Connection String
if TRUSTED_CONNECTION == 'yes':
    conn_str = f'mssql+pyodbc://@{SERVER}/{DATABASE}?driver={DRIVER}&trusted_connection=yes'
else:
    conn_str = f'mssql+pyodbc://{USERNAME}:{PASSWORD}@{SERVER}/{DATABASE}?driver={DRIVER}'

print("Cadena de conexi√≥n construida (oculta password):", conn_str.split('@')[-1])
```

## Paso 3: Probar la Conexi√≥n
Intentaremos conectar y ejecutar una consulta simple.

```python
try:
    engine = sqlalchemy.create_engine(conn_str)
    connection = engine.connect()
    print("‚úÖ ¬°Conexi√≥n exitosa!")
    
    # Prueba r√°pida
    df = pd.read_sql("SELECT @@VERSION as version", connection)
    print("\nVersi√≥n de SQL Server:")
    print(df['version'][0])
    
    connection.close()
except Exception as e:
    print("‚ùå Error de conexi√≥n:")
    print(e)
```
